# Some blueprint instructions for an aspiring developer

Always be on the `orchestration` branch of cstar till further notice.

The schema for a blueprint is defined in the [RomsMarblBlueprint](https://github.com/CWorthy-ocean/C-Star/blob/orchestration/cstar/orchestration/models.py#L319) class and has several nested substructures

If you want to just check that you've made a valid blueprint, you can do something like:
```python
from cstar.orchestration.models import RomsMarblBlueprint
from cstar.orchestration.serialization import deserialize
model = deserialize(path_to_your_blueprint_yaml, RomsMarblBlueprint)  # will raise error if invalid
```

There are a number of ways you could potentially run a simulation from a blueprint and some of them are a little fraught at the moment. 

If I were doing it within a notebook and running something small on a Mac, I think I'd do something like:
```python
from cstar.roms import ROMSSimulation
sim = ROMSSimulation.from_blueprint(path_to_your_blueprint_yaml)
sim.setup()    # stage code and data, build roms executable
sim.pre_run()  # partition, if needed
sim.run()      # run roms itself
sim.post_run() # join files, note that this will do some auto-parallelization and could get you kicked off a login node
```
